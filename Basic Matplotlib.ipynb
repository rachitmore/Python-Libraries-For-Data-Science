```python
# Importing Matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Simple Line Plot

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a plot
plt.plot(x, y)

# Add a title and labels
plt.title("Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()
```

```python
# Scatter Plot

# Sample data
x = np.random.rand(100)
y = np.random.rand(100)

# Create a scatter plot
plt.scatter(x, y)

# Add a title and labels
plt.title("Random Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()
```

```python
# Bar Plot

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 9]

# Create a bar plot
plt.bar(categories, values)

# Add a title and labels
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")

# Show the plot
plt.show()
```

```python
# Histogram

# Sample data
data = np.random.randn(1000)

# Create a histogram
plt.hist(data, bins=30)

# Add a title and labels
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

```python
# Pie Chart

# Sample data
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Add a title
plt.title("Pie Chart Example")

# Show the plot
plt.show()
```

```python
# Subplots

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1)

# First subplot
ax1.plot(x, y1)
ax1.set_title("Sine Wave")

# Second subplot
ax2.plot(x, y2)
ax2.set_title("Cosine Wave")

# Add a title to the figure
fig.suptitle("Subplot Example")

# Show the plot
plt.show()
```

```python
# Customizing Plots

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a plot
plt.plot(x, y, color='green', linestyle='--', linewidth=2, marker='o', markersize=5)

# Add a title and labels
plt.title("Customized Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()
```

```python
# Saving Plots

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a plot
plt.plot(x, y)

# Save the plot
plt.savefig("sine_wave.png")

# Show the plot
plt.show()
```

```python
# Adding Legends

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a plot
plt.plot(x, y1, label='Sine')
plt.plot(x, y2, label='Cosine')

# Add a legend
plt.legend()

# Add a title and labels
plt.title("Sine and Cosine Waves")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot
plt.show()
```
