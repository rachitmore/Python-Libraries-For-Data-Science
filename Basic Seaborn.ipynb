```python
# Importing Seaborn and Matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Scatter Plot

import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
tips = sns.load_dataset("tips")

# Create a scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip")

# Show the plot
plt.show()
```

```python
# Line Plot

# Load the dataset
fmri = sns.load_dataset("fmri")

# Create a line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Show the plot
plt.show()
```

```python
# Bar Plot

# Create a bar plot
sns.barplot(data=tips, x="day", y="total_bill", ci=None)

# Show the plot
plt.show()
```

```python
# Histogram

# Create a histogram
sns.histplot(data=tips, x="total_bill", bins=20, kde=True)

# Show the plot
plt.show()
```

```python
# Box Plot

# Create a box plot
sns.boxplot(data=tips, x="day", y="total_bill")

# Show the plot
plt.show()
```

```python
# Pair Plot

# Create a pair plot
sns.pairplot(tips)

# Show the plot
plt.show()
```

```python
# Heatmap

# Load the dataset
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

# Create a heatmap
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")

# Show the plot
plt.show()
```

```python
# FacetGrid

# Create a FacetGrid
g = sns.FacetGrid(tips, col="sex", hue="smoker")
g.map(plt.scatter, "total_bill", "tip", alpha=.7)
g.add_legend()

# Show the plot
plt.show()
```

```python
# Joint Plot

# Create a joint plot
sns.jointplot(data=tips, x="total_bill", y="tip", kind="reg")

# Show the plot
plt.show()
```

```python
# Violin Plot

# Create a violin plot
sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker", split=True)

# Show the plot
plt.show()
```

```python
# Swarm Plot

# Create a swarm plot
sns.swarmplot(data=tips, x="day", y="total_bill", hue="smoker", dodge=True)

# Show the plot
plt.show()
```
